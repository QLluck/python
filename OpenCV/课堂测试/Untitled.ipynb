```python
import cv2
import numpy as np
```

```python
img = cv2.imread("car.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
```

```python
ret_fixed, binary_fixed = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
```

```python
ret_otsu, binary_otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
import matplotlib.pyplot as plt

# 绘图展示
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.imshow(gray, cmap="gray")
plt.title("Gray Image")
plt.axis("off")

plt.subplot(132)
plt.imshow(binary_fixed, cmap="gray")
plt.title("Fixed Threshold")
plt.axis("off")

plt.subplot(133)
plt.imshow(binary_otsu, cmap="gray")
plt.title("Otsu Threshold")
plt.axis("off")

plt.tight_layout()
plt.show()
```
