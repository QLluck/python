# 聚类算法实验

**实验目标**：掌握 K-means、DBSCAN、层次聚类三种算法，对比其在不同类型数据集上的性能差异。

## 1. 环境准备

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from sklearn.datasets import make_blobs, make_moons, make_circles
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, calinski_harabasz_score
from scipy.cluster.hierarchy import dendrogram, linkage
import os, sys, warnings
warnings.filterwarnings('ignore')

# 中文字体设置
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 创建保存图片的目录
os.makedirs('./figures', exist_ok=True)

print(f"Python: {sys.version.split()[0]}")
for lib in [np, matplotlib]:
    print(f"{lib.__name__}: {lib.__version__}")
import sklearn; print(f"sklearn: {sklearn.__version__}")
import scipy; print(f"scipy: {scipy.__version__}")

RANDOM_STATE = 42
```

## 2. 数据准备

### 2.1 生成合成数据集

使用 sklearn 生成三种不同分布的数据集：
- **高斯分布数据** (`make_blobs`)：球状簇，适合 K-means
- **月牙形数据** (`make_moons`)：非凸形状，挑战 K-means
- **环形数据** (`make_circles`)：同心圆，K-means 难以处理

```python
n_samples = 500

# 高斯分布数据
X_blobs, y_blobs_true = make_blobs(
    n_samples=n_samples, centers=4, cluster_std=1.5,
    random_state=RANDOM_STATE
)

# 月牙形数据
X_moons, y_moons_true = make_moons(
    n_samples=n_samples, noise=0.08, random_state=RANDOM_STATE
)

# 环形数据
X_circles, y_circles_true = make_circles(
    n_samples=n_samples, noise=0.05, factor=0.5, random_state=RANDOM_STATE
)

datasets = {
    'Blobs (高斯分布)': (X_blobs, y_blobs_true),
    'Moons (月牙形)': (X_moons, y_moons_true),
    'Circles (环形)': (X_circles, y_circles_true),
}

# 可视化原始数据
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, (name, (X, y)) in zip(axes, datasets.items()):
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap='Set2', s=15, alpha=0.8)
    ax.set_title(f'{name}\n(真实标签)', fontsize=13)
    ax.set_xlabel('x'); ax.set_ylabel('y')
plt.tight_layout()
plt.savefig('./figures/raw_datasets.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 2.2 数据预处理：Z-score 标准化

使用 `StandardScaler` 将数据标准化为均值 0、标准差 1 的分布。

```python
# 标准化
scaler_blobs = StandardScaler().fit(X_blobs)
scaler_moons  = StandardScaler().fit(X_moons)
scaler_circles = StandardScaler().fit(X_circles)

X_blobs_scaled = scaler_blobs.transform(X_blobs)
X_moons_scaled  = scaler_moons.transform(X_moons)
X_circles_scaled = scaler_circles.transform(X_circles)

datasets_scaled = {
    'Blobs (高斯分布)': X_blobs_scaled,
    'Moons (月牙形)': X_moons_scaled,
    'Circles (环形)': X_circles_scaled,
}

# 可视化标准化前后的数据分布
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
for j, (name, (X_raw, _)) in enumerate(datasets.items()):
    axes[0, j].scatter(X_raw[:, 0], X_raw[:, 1], s=10, alpha=0.6)
    axes[0, j].set_title(f'{name} (原始)')
    axes[0, j].set_xlabel('x'); axes[0, j].set_ylabel('y')
    X_sc = datasets_scaled[name]
    axes[1, j].scatter(X_sc[:, 0], X_sc[:, 1], s=10, alpha=0.6)
    axes[1, j].set_title(f'{name} (标准化后)')
    axes[1, j].set_xlabel('x'); axes[1, j].set_ylabel('y')
plt.suptitle('标准化前后对比', fontsize=14)
plt.tight_layout()
plt.savefig('./figures/scaling_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. K-means 聚类

### 3.1 肘部法则选择最佳 K 值

对每个数据集，计算 K=1~10 的 SSE（簇内误差平方和），找到肘部位置。

```python
def plot_elbow(X, dataset_name):
    """绘制肘部法则曲线"""
    sse = []
    K_range = range(1, 11)
    for k in K_range:
        km = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init='auto')
        km.fit(X)
        sse.append(km.inertia_)

    plt.figure(figsize=(6, 4))
    plt.plot(K_range, sse, 'bo-', markersize=6)
    plt.xlabel('K'); plt.ylabel('SSE')
    plt.title(f'{dataset_name} - 肘部法则')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    safe_name = dataset_name.replace(' ', '_').replace('(', '').replace(')', '')
    fname = f'./figures/elbow_{safe_name}.png'
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()

    # 最大曲率法找肘部
    diffs = np.diff(sse)
    diffs2 = np.diff(diffs)
    best_k = np.argmax(diffs2) + 2
    print(f"  推荐 K = {best_k}")
    return best_k

best_k_dict = {}
for name, X in datasets_scaled.items():
    print(f"\n{'='*40}\n{name}\n{'='*40}")
    best_k_dict[name] = plot_elbow(X, name)
```

### 3.2 K-means 聚类结果可视化

```python
def run_kmeans(X, dataset_name, k):
    """K-means 聚类并可视化"""
    km = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init='auto')
    labels = km.fit_predict(X)
    sil = silhouette_score(X, labels)
    ch = calinski_harabasz_score(X, labels)

    plt.figure(figsize=(6, 4))
    for lbl in np.unique(labels):
        mask = labels == lbl
        plt.scatter(X[mask, 0], X[mask, 1], s=15, alpha=0.7, label=f'Cluster {lbl}')
    centers = km.cluster_centers_
    plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200,
                edgecolors='black', linewidths=1.5, label='Centers', zorder=5)
    plt.title(f'{dataset_name}\nK={k} | Silhouette={sil:.3f} | CH={ch:.1f}')
    plt.xlabel('x'); plt.ylabel('y')
    plt.legend(fontsize=7, loc='best')
    plt.tight_layout()
    safe_name = dataset_name.replace(' ', '_').replace('(', '').replace(')', '')
    fname = f'./figures/kmeans_{safe_name}.png'
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()
    return labels, sil, ch

kmeans_results = {}
for name, X in datasets_scaled.items():
    k = best_k_dict[name]
    labels, sil, ch = run_kmeans(X, name, k)
    kmeans_results[name] = {'labels': labels, 'silhouette': sil, 'ch': ch, 'k': k}
```

## 4. DBSCAN 密度聚类

DBSCAN 通过密度连通性发现任意形状的簇。关键参数：
- **eps**：邻域半径
- **min_samples**：核心点的最小邻居数

对每个数据集尝试多组参数组合，选择轮廓系数最高的结果。

```python
def run_dbscan_grid(X, dataset_name, eps_list=None, min_samples_list=None):
    """遍历 DBSCAN 参数网格，返回最优结果"""
    if eps_list is None:
        eps_list = [0.1, 0.2, 0.3, 0.5, 0.8, 1.0]
    if min_samples_list is None:
        min_samples_list = [3, 5, 10, 15]

    best_sil = -1
    best_result = None
    results = []

    for eps in eps_list:
        for ms in min_samples_list:
            db = DBSCAN(eps=eps, min_samples=ms)
            labels = db.fit_predict(X)
            n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
            n_noise = np.sum(labels == -1)
            if n_clusters >= 2:
                sil = silhouette_score(X, labels)
                ch = calinski_harabasz_score(X, labels)
            else:
                sil, ch = -1, 0
            results.append((eps, ms, n_clusters, n_noise, sil, ch, labels))
            if sil > best_sil:
                best_sil = sil
                best_result = (eps, ms, n_clusters, n_noise, sil, ch, labels)

    print(f"\n  {'eps':<8} {'ms':<6} {'clusters':<8} {'noise':<6} {'Silhouette':<12} {'CH':<10}")
    print("  " + "-"*52)
    results_sorted = sorted(results, key=lambda x: x[4], reverse=True)
    for eps, ms, nc, nn, sil, ch, _ in results_sorted[:5]:
        print(f"  {eps:<8} {ms:<6} {nc:<8} {nn:<6} {sil:<12.3f} {ch:<10.1f}")
    return best_result

dbscan_results = {}
for name, X in datasets_scaled.items():
    print(f"\n{'='*50}\n{name}\n{'='*50}")
    best = run_dbscan_grid(X, name)
    eps, ms, nc, nn, sil, ch, labels = best
    dbscan_results[name] = {'labels': labels, 'silhouette': sil, 'ch': ch,
                              'eps': eps, 'min_samples': ms, 'n_clusters': nc, 'n_noise': nn}
    print(f"  Best: eps={eps}, min_samples={ms}, clusters={nc}, noise={nn}")
```

### 4.1 DBSCAN 聚类结果可视化

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, (name, X) in zip(axes, datasets_scaled.items()):
    res = dbscan_results[name]
    labels = res['labels']
    unique_labels = np.unique(labels)
    colors = plt.cm.Set2(np.linspace(0, 1, len(unique_labels)))
    for lbl, color in zip(unique_labels, colors):
        mask = labels == lbl
        label_name = 'Noise' if lbl == -1 else f'Cluster {lbl}'
        marker = 'x' if lbl == -1 else 'o'
        ax.scatter(X[mask, 0], X[mask, 1], c=[color], s=15 if lbl != -1 else 25,
                   marker=marker, alpha=0.7, label=label_name)
    ax.set_title(f'{name}\neps={res["eps"]}, ms={res["min_samples"]} | Sil={res["silhouette"]:.3f}')
    ax.set_xlabel('x'); ax.set_ylabel('y')
    ax.legend(fontsize=6, loc='best')
plt.tight_layout()
plt.blob('./figures/dbscan_all.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. 层次聚类 (Agglomerative Clustering)

尝试不同的连接方式（linkage）：
- **Ward**：最小化簇内方差增量
- **Average**：簇间平均距离
- **Complete**：簇间最大距离

```python
def run_agg(X, dataset_name, n_clusters, linkage='ward'):
    """层次聚类并返回指标"""
    agg = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage)
    labels = agg.fit_predict(X)
    sil = silhouette_score(X, labels) if len(set(labels)) > 1 else -1
    ch = calinski_harabasz_score(X, labels) if len(set(labels)) > 1 else 0
    return labels, sil, ch

linkage_options = ['ward', 'average', 'complete']
agg_best_results = {}

for name, X in datasets_scaled.items():
    print(f"\n{'='*50}\n{name}\n{'='*50}")
    best_sil = -1
    best_res = None
    k = best_k_dict[name]
    for link in linkage_options:
        labels, sil, ch = run_agg(X, name, k, link)
        print(f"  linkage={link:<8} -> Silhouette={sil:.3f}, CH={ch:.1f}")
        if sil > best_sil:
            best_sil = sil
            best_res = (labels, sil, ch, link)
    agg_best_results[name] = {
        'labels': best_res[0], 'silhouette': best_res[1],
        'ch': best_res[2], 'linkage': best_res[3]
    }
    print(f"  Best linkage = {best_res[3]}")
```

### 5.1 层次聚类结果可视化

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, (name, X) in zip(axes, datasets_scaled.items()):
    res = agg_best_results[name]
    labels = res['labels']
    for lbl in np.unique(labels):
        mask = labels == lbl
        ax.scatter(X[mask, 0], X[mask, 1], s=15, alpha=0.7, label=f'Cluster {lbl}')
    ax.set_title(f'{name}\nlinkage={res["linkage"]} | Sil={res["silhouette"]:.3f}')
    ax.set_xlabel('x'); ax.set_ylabel('y')
    ax.legend(fontsize=7, loc='best')
plt.tight_layout()
plt.savefig('./figures/agg_all.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 5.2 树状图 (Dendrogram)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for ax, (name, X) in zip(axes, datasets_scaled.items()):
    Z = linkage(X, method='ward')
    dendrogram(Z, ax=ax, truncate_mode='lastp', p=30,
               leaf_rotation=0, leaf_font_size=8)
    ax.set_title(f'{name} - Dendrogram (Ward)')
    ax.set_xlabel('Sample Index'); ax.set_ylabel('Distance')
plt.tight_layout()
plt.savefig('./figures/dendrograms.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 6. 性能对比汇总

汇总三种算法在三种数据集上的轮廓系数和 Calinski-Harabasz 指数。

```python
import pandas as pd

rows = []
for name in datasets_scaled.keys():
    rows.append({
        'Dataset': name,
        'K-means Silhouette': f"{kmeans_results[name]['silhouette']:.3f}",
        'K-means CH': f"{kmeans_results[name]['ch']:.1f}",
        'DBSCAN Silhouette': f"{dbscan_results[name]['silhouette']:.3f}",
        'DBSCAN CH': f"{dbscan_results[name]['ch']:.1f}",
        'Agglomerative Silhouette': f"{agg_best_results[name]['silhouette']:.3f}",
        'Agglomerative CH': f"{agg_best_results[name]['ch']:.1f}",
    })

df = pd.DataFrame(rows)
print(df.to_string(index=False))
```

### 6.1 结果分析

- **高斯分布数据 (Blobs)**：K-means 和层次聚类表现最佳，因为数据呈球状分布；DBSCAN 在合适参数下也能取得良好效果
- **月牙形数据 (Moons)**：DBSCAN 能识别非凸形状，表现优于 K-means；层次聚类的不同 linkage 对结果影响较大
- **环形数据 (Circles)**：DBSCAN 优势最明显，能准确识别同心圆结构；K-means 基于距离划分，无法处理此类数据

### 6.2 算法适用场景总结

| 算法 | 优点 | 缺点 | 适用场景 |
|------|------|------|----------|
| K-means | 简单高效，可解释性强 | 需预设K，只能发现球状簇 | 数据分布近似球形、已知类别数 |
| DBSCAN | 可发现任意形状簇，自动识别噪声 | 参数敏感，密度不均时效果差 | 空间聚类、异常检测、非凸形状 |
| 层次聚类 | 无需预设簇数，树状图直观 | 计算复杂度高，不可逆 | 小规模数据、需要层次关系 |
