```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

# ==========================
# 任务一：基础预处理
# ==========================
def task1_preprocess(img_path):
    print("===== 任务一：图像基础预处理 =====")
    
    # 1. 读取图像
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # 2. 灰度化
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    print("已完成灰度化")

    # 3. 截取感兴趣区域 ROI（车牌大致位置，可自行调整）
    h, w = gray.shape
    roi = gray[int(h*0.3):int(h*0.7), int(w*0.2):int(w*0.8)]
    print("已完成 ROI 截取")

    # 4. 计算直方图
    hist = cv2.calcHist([roi], [0], None, [256], [0,256])
    print("已完成直方图计算")

    # 5. 图像平滑去噪
    blur = cv2.GaussianBlur(roi, (5,5), 0)  # 高斯滤波
    
    # 显示结果
    # Show results
    plt.figure(figsize=(12,8))
    plt.subplot(221), plt.imshow(img_rgb), plt.title('Original Image')
    plt.subplot(222), plt.imshow(gray, 'gray'), plt.title('Grayscale Image')
    plt.subplot(223), plt.imshow(roi, 'gray'), plt.title('ROI')
    plt.subplot(224), plt.plot(hist), plt.title('Histogram')
    plt.show()
    
    plt.imshow(blur, 'gray'), plt.title('Gaussian Smoothing & Denoising')
    plt.show()
    return gray, roi, blur
```

```python

# ==========================
# 任务二：图像增强 + 质量分析
# ==========================
def task2_enhance(gray):
    print("\n===== 任务二：图像增强与质量分析 =====")
    
    # 1. 均值滤波
    mean_blur = cv2.blur(gray, (5,5))
    
    # 2. 中值滤波（去椒盐噪声最优）
    median_blur = cv2.medianBlur(gray, 5)
    
    # 3. 高斯滤波
    gauss_blur = cv2.GaussianBlur(gray, (5,5), 0)
    
    # 4. 自适应直方图均衡化 CLAHE
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    clahe_img = clahe.apply(gray)
    
    # 显示效果
    plt.figure(figsize=(12,8))
    plt.subplot(221), plt.imshow(mean_blur,'gray'), plt.title('Mean Filter')
    plt.subplot(222), plt.imshow(median_blur,'gray'), plt.title('Median Filter')
    plt.subplot(223), plt.imshow(gauss_blur,'gray'), plt.title('Gaussian Filter')
    plt.subplot(224), plt.imshow(clahe_img,'gray'), plt.title('CLAHE Enhancement')
    plt.show()
    
    print("完成：滤波去噪 + CLAHE对比度增强")
    print("优化建议：增大核(7,7)去噪更强，但会模糊；clipLimit改3~4对比度更高")
    return clahe_img
```

```python
# ==========================
# 任务三：傅里叶变换 + 频域滤波
# ==========================
def task3_fourier_filter(gray):
    print("\n===== 任务三：傅里叶变换与频域滤波 =====")
    
    # 1. 傅里叶变换
    fft = np.fft.fft2(gray)
    fft_shift = np.fft.fftshift(fft)
    magnitude = 20 * np.log(np.abs(fft_shift))
    

    # 2. 低通滤波器（去噪）
    h, w = gray.shape
    crow, ccol = h//2, w//2
    mask_low = np.zeros((h,w), np.uint8)
    cv2.circle(mask_low, (ccol,crow), 50, 1, -1)

    # 3. 高通滤波器（锐化/边缘）
    mask_high = np.ones((h,w), np.uint8)
    cv2.circle(mask_high, (ccol,crow), 20, 0, -1)

    # 4. 滤波 + 逆变换
    # 低通
    fft_low = fft_shift * mask_low
    img_low = np.fft.ifft2(np.fft.ifftshift(fft_low))
    img_low = np.abs(img_low)
    
    # 高通
    fft_high = fft_shift * mask_high
    img_high = np.fft.ifft2(np.fft.ifftshift(fft_high))
    img_high = np.abs(img_high)

    # 显示
    plt.figure(figsize=(12,8))
    plt.subplot(221), plt.imshow(magnitude, 'gray'), plt.title('Magnitude Spectrum')
    plt.subplot(222), plt.imshow(mask_low, 'gray'), plt.title('Low-pass Mask')
    plt.subplot(223), plt.imshow(img_low, 'gray'), plt.title('Low-pass Denoising')
    plt.subplot(224), plt.imshow(img_high, 'gray'), plt.title('High-pass Edge')
    plt.show()
```

```python


# ==========================
# 主函数：运行全部实验
# ==========================
if __name__ == "__main__":
    img_path = "car.jpg"  # 你的车牌图片名
    gray, roi, blur = task1_preprocess(img_path)
    enhance_img = task2_enhance(gray)
    task3_fourier_filter(gray)
```
