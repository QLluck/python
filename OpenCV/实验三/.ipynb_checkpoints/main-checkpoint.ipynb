```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 统一绘图函数
def show_images(images, titles, cols=3):
    plt.figure(figsize=(15, 10))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot((len(images) + cols - 1) // cols, cols, i+1)
        if len(img.shape) == 3:
            plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        else:
            plt.imshow(img, cmap='gray')
        plt.title(title)
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
# 1. 读取并二值化示例图像（可换成你自己的图片路径）
img = cv2.imread("test_image.jpg", cv2.IMREAD_GRAYSCALE)
# 转为二值图（Otsu自动阈值）
_, binary_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 2. 定义不同结构元素
kernel_rect = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))   # 矩形
kernel_cross = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5)) # 十字形
kernel_ellipse = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5)) # 圆形

# 3. 腐蚀与膨胀（以矩形核为例）
erosion = cv2.erode(binary_img, kernel_rect, iterations=1)
dilation = cv2.dilate(binary_img, kernel_rect, iterations=1)

# 4. 开运算与闭运算
opening = cv2.morphologyEx(binary_img, cv2.MORPH_OPEN, kernel_rect)  # 先腐蚀后膨胀
closing = cv2.morphologyEx(binary_img, cv2.MORPH_CLOSE, kernel_rect)  # 先膨胀后腐蚀

# 显示结果
show_images(
    [binary_img, erosion, dilation, opening, closing],
    ["原始二值图", "腐蚀", "膨胀", "开运算", "闭运算"]
)
```

```python
# 1. 读取带噪声的图像（如你提供的带噪文字/指纹图）
noisy_img = cv2.imread("noisy_image.jpg", cv2.IMREAD_GRAYSCALE)
_, binary_noisy = cv2.threshold(noisy_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 2. 用不同形态学操作去噪
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
# 开运算：去除小的亮噪声
denoise_open = cv2.morphologyEx(binary_noisy, cv2.MORPH_OPEN, kernel)
# 闭运算：去除小的暗噪声
denoise_close = cv2.morphologyEx(binary_noisy, cv2.MORPH_CLOSE, kernel)
# 开+闭组合：同时处理亮、暗噪声
denoise_openclose = cv2.morphologyEx(denoise_open, cv2.MORPH_CLOSE, kernel)

# 显示结果
show_images(
    [binary_noisy, denoise_open, denoise_close, denoise_openclose],
    ["原始带噪图", "开运算去噪", "闭运算去噪", "开+闭去噪"]
)                                                                                      
```

```python
# 1. 读取灰度图像
img = cv2.imread("edge_test.png", cv2.IMREAD_GRAYSCALE)
img = cv2.GaussianBlur(img, (3, 3), 0)  # 先平滑去噪，减少伪边缘

# 2. 形态学梯度边缘检测
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)  # 膨胀-腐蚀

# 3. 拉普拉斯空域边缘检测
laplacian = cv2.Laplacian(img, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)

# 4. 频域巴特沃斯高通滤波边缘检测
def butterworth_highpass(img, d0, n):
    rows, cols = img.shape
    crow, ccol = rows//2, cols//2
    # 傅里叶变换
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    # 构造高通滤波器
    u, v = np.meshgrid(np.arange(cols), np.arange(rows))
    d = np.sqrt((u - ccol)**2 + (v - crow)**2)
    h = 1 / (1 + (d0 / (d + 1e-8))**(2*n))
    # 滤波
    fshift_filtered = fshift * h
    # 逆变换
    f_ishift = np.fft.ifftshift(fshift_filtered)
    img_back = np.fft.ifft2(f_ishift)
    img_back = np.abs(img_back)
    # 归一化
    img_back = cv2.normalize(img_back, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
    return img_back

bw_highpass = butterworth_highpass(img, d0=30, n=2)

# 显示对比
show_images(
    [img, gradient, laplacian, bw_highpass],
    ["原始图像", "形态学梯度", "拉普拉斯", "巴特沃斯高通"]
)
```

```python
# 1. 读取指纹图像
fingerprint = cv2.imread("fingerprint.png", cv2.IMREAD_GRAYSCALE)
# 预处理：自适应二值化，处理光照不均
binary_fp = cv2.adaptiveThreshold(
    fingerprint, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2
)

# 2. 形态学滤波增强指纹细节
# 先开运算去除背景噪声
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
fp_open = cv2.morphologyEx(binary_fp, cv2.MORPH_OPEN, kernel)
# 再闭运算填充纹路断裂
fp_enhanced = cv2.morphologyEx(fp_open, cv2.MORPH_CLOSE, kernel)

# 3. 可选：细化指纹（骨架提取）
def skeletonize(img):
    img = img.copy()
    size = np.size(img)
    skel = np.zeros(img.shape, np.uint8)
    element = cv2.getStructuringElement(cv2.MORPH_CROSS, (3,3))
    while True:
        eroded = cv2.erode(img, element)
        temp = cv2.dilate(eroded, element)
        temp = cv2.subtract(img, temp)
        skel = cv2.bitwise_or(skel, temp)
        img = eroded.copy()
        zeros = size - cv2.countNonZero(img)
        if zeros == size:
            break
    return skel

fp_skeleton = skeletonize(fp_enhanced)

# 显示结果
show_images(
    [fingerprint, binary_fp, fp_enhanced, fp_skeleton],
    ["原始指纹", "自适应二值化", "形态学增强", "指纹骨架"]
)
```
