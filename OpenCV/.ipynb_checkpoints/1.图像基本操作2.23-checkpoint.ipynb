# 数据读取-视频

cv2.VideoCapture可以捕获摄像头，用数字来控制不同的设备，例如0,1。

如果是视频文件，直接指定好路径即可。

```python
#1、实时捕捉摄像头数据,如果没有摄像头或读不到数据会闪退，不报错
import cv2
cv2.namedWindow('video',cv2.WINDOW_NORMAL)
cv2.resizeWindow('video',640,480)
cap = cv2.VideoCapture(0)
#循环读取摄像头的每一帧
while True: #while cap.isOened():
    #读一帧数据，返回标记和这一帧数据，True表示读到了数据，False表示没读到数据
    ret,frame=cap.read()
    #可以根据ret做个判断
    if not ret:
        #没读到数据
        break
    #显示数据
    cv2.imshow('video',frame)
    key=cv2.waitKey(0)
    if key & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

# 案例2-2 彩色图像转灰度图像

```python
#1、导入包
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#2、读入彩色图像
rgbImg = cv2.imread("C:/Users/Administrator/Desktop/img/lena.jpg")
```

```python
#3、获取图像的宽、高信息
height = rgbImg.shape[0]
width = rgbImg.shape[1]
height = rgbImg.shape[0]
width = rgbImg.shape[1]
```

```python
#4、初始化三张单通道图像作为目标图像，图像宽高跟彩色图像保持一致
grayImg_maxVal = np.zeros([height,width],np.uint8)
grayImg_meanVal = np.zeros([height,width],np.uint8)
grayImg_w_meanVal = np.zeros([height,width],np.uint8)
```

```python
#5、获取彩色图像每个坐标位置处的r，g，b三个通道的像素值，并分别取三通道像素值的最大值、平均值和加权平均值写入目标图像
for row in range(height):
    for col in range(width):
        r_value = rgbImg[row,col,2]
        g_value = rgbImg[row,col,1]
        b_value = rgbImg[row,col,0]
        value1 = max(r_value,g_value,b_value)
        value2 = r_value//3 + g_value//3 + b_value//3
        value3 = r_value*0.3 + g_value*0.4 + b_value*0.3
        grayImg_maxVal[row,col] = value1
        grayImg_meanVal[row,col] = value2
        grayImg_w_meanVal[row,col] = value3
```

```python
#6、显示结果
#1.最大值法
plt.imshow(grayImg_maxVal,cmap = 'gray')
```

```python
#2.平均值法
plt.imshow(grayImg_meanVal,cmap = 'gray')
```

```python
#3.加权平均值法
plt.imshow(grayImg_w_meanVal,cmap = 'gray')
```

# trackbar的用法

```python
import cv2
import numpy as np
#创建窗口
cv2.namedWindow('trackbar',cv2.WINDOW_NORMAL)
cv2.resizeWindow('trackbar',640,480)
#定义回调函数
def callback(value):
    #print(value)
    pass
#创建3个trackbar
cv2.createTrackbar('R','trackbar',0,255,callback)#需要callback函数
cv2.createTrackbar('G','trackbar',0,255,callback)
cv2.createTrackbar('B','trackbar',0,255,callback)
#创建背景图片
img = np.zeros((480,640,3),np.uint8)
while True:
    #获取当前trackbar的值
    r = cv2.getTrackbarPos('R','trackbar')
    g = cv2.getTrackbarPos('G','trackbar')
    b = cv2.getTrackbarPos('B','trackbar')
    #用获取到的三个值修改背景图片颜色
    img[:] = [b,g,r]
    cv2.imshow('trackbar',img)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
cv2.destroyAllWindows()
```

# 颜色空间转换

```python
#cv2.cvtColor()
import cv2
def callback(value):
    pass
cv2.namedWindow('color',cv2.WINDOW_NORMAL)
cv2.resizeWindow('color',640,480)
#读取图片，默认BGR
img = cv2.imread("C:/Users/Administrator/Desktop/img/cat.jpg")
#定义颜色控件转化列表
color_spaces = [cv2.COLOR_BGR2RGB,cv2.COLOR_BGR2RGBA,cv2.COLOR_BGR2GRAY,cv2.COLOR_BGR2HSV,cv2.COLOR_BGR2YUV]
#设置trackbar
cv2.createTrackbar('trackbar','color',0,4,callback)
while True:
    #获取trackbar的值
    index = cv2.getTrackbarPos('trackbar','color')
    #进行颜色空间转换
    cvt_img = cv2.cvtColor(img,color_spaces[index])
    cv2.imshow('color',cvt_img)
    key = cv2.waitKey(10)
    if key == ord('q'):
        break
cv2.destroyAllWindows()
    
```

# 图像算术运算

```python
#1.加法
import cv2
import numpy as np

cat=cv2.imread(r'C:\Users\Administrator\Desktop\img\cat1.jpg')
dog=cv2.imread(r'C:\Users\Administrator\Desktop\img\dog.jpg')
cat_dog=cv2.add(cat,dog)
cv2.imshow('cat and dog',np.hstack((cat,dog)))
cv2.imshow('cat+dog',cat_dog)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

注意： 
1. 图片就是矩阵，图片的加法运算就是矩阵的加法运算，这就要求加法运算的两张图像的shape必须是相同的。

2. add() 加法操作就是将两个图像对应位置的元素相加，如果超过255，则全部变成255。

3. 图像还可以和单个数字进行运算

```python
print(cat[:2,:2])
cat += 100
cv2.imshow('cat+100',cat)
print(cat[:2,:2])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 2.减法
cat=cv2.imread(r'C:\Users\Administrator\Desktop\img\cat1.jpg')
dog=cv2.imread(r'C:\Users\Administrator\Desktop\img\dog.jpg')
cat_dog=cv2.subtract(cat,dog)
cv2.imshow('cat-dog',cat_dog)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

对应位置的元素相减，减完若小于0，则统一变成0。

```python
print(cat[:2,:2])
print(dog[:2,:2])
print(cat_dog[:2,:2])
```

```python
# 3.乘法
cat=cv2.imread(r'C:\Users\Administrator\Desktop\img\cat1.jpg')
dog=cv2.imread(r'C:\Users\Administrator\Desktop\img\dog.jpg')
cat_dog=cv2.multiply(cat,dog)
cv2.imshow('cat*dog',cat_dog)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 4.除法
cat=cv2.imread(r'C:\Users\Administrator\Desktop\img\cat1.jpg')
dog=cv2.imread(r'C:\Users\Administrator\Desktop\img\dog.jpg')
cat_dog=cv2.divide(cat,dog)
cv2.imshow('cat/dog',cat_dog)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 图像融合

# 图像位运算

```python
# 1.非运算
import cv2
import numpy as np
#全黑图片
img1=np.zeros((480,640),np.uint8)
#部分设置成白色
img1[120:360,160:480]=255
img_not=cv2.bitwise_not(img1)
cv2.imshow('img',np.hstack((img1,img_not)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 2.与运算
#两张全黑图片
img1=np.zeros((480,640),np.uint8)
img2=np.zeros((480,640),np.uint8)
#部分设置成白色
img1[0:360,0:480]=255
img2[120:480,160:480]=255
img_and=cv2.bitwise_and(img1,img2)
cv2.imshow('img1 img2',np.hstack((img1,img2)))
cv2.imshow('img_and',img_and)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 3.或运算
#两张全黑图片
img1=np.zeros((480,640),np.uint8)
img2=np.zeros((480,640),np.uint8)
#部分设置成白色
img1[0:360,0:480]=255
img2[120:480,160:480]=255
img_or=cv2.bitwise_or(img1,img2)
cv2.imshow('img1 img2',np.hstack((img1,img2)))
cv2.imshow('img_or',img_or)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 4.异或运算
#两张全黑图片
img1=np.zeros((480,640),np.uint8)
img2=np.zeros((480,640),np.uint8)
#部分设置成白色
img1[0:360,0:480]=255
img2[120:480,160:480]=255
img_xor=cv2.bitwise_xor(img1,img2)
cv2.imshow('img1 img2',np.hstack((img1,img2)))
cv2.imshow('img_xor',img_xor)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 制作LOGO

```python
import cv2
import numpy as np
#导入图像
dog = cv2.imread(r'C:\Users\Administrator\Desktop\img\dog.jpg')
#创建LOGO
logo=np.zeros((200,200,3),np.uint8)
#绘制LOGO
logo[20:120,20:120]=[0,0,255]
logo[80:180,80:180]=[0,255,0]
#掩码
mask=np.zeros((200,200),np.uint8)
mask[20:120,20:120]=255
mask[80:180,80:180]=255
cv2.imshow('mask',mask)
m=cv2.bitwise_not(mask)
cv2.imshow('m',m)
#确定dog添加logo的位置
roi=dog[0:200,0:200]
#roi与dog进行与操作，先roi和roi做与运算，然后结果再和mask做与运算，如果与的结果为True，那么返回原图的像素，否则返回0
tmp=cv2.bitwise_and(roi,roi,mask=m)
cv2.imshow('tmp',tmp)
dst=cv2.add(tmp,logo)
cv2.imshow('dst',dst)
#在dog上还原
dog[:200,:200]=dst
cv2.imshow('dog',dog)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 绘制图形

```python
import cv2
import numpy as np
#创建纯黑的背景用来画图形
img = np.zeros((480,640,3),np.uint8)
#1、画线cv2.line(img,pt1,pt2,color,thickness,lineType,shift)
cv2.line(img,(10,20),(300,400),(0,0,255),5,480)
cv2.imshow('draw',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 图像边界填充

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread(r'C:\Users\Administrator\Desktop\img\cat.jpg')
top_size,bottom_size,left_size,right_size = (50,50,50,50)

replicate = cv2.copyMakeBorder(img, top_size, bottom_size, left_size, right_size, borderType=cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img, top_size, bottom_size, left_size, right_size,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img, top_size, bottom_size, left_size, right_size, cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img, top_size, bottom_size, left_size, right_size, cv2.BORDER_WRAP)
constant = cv2.copyMakeBorder(img, top_size, bottom_size, left_size, right_size,cv2.BORDER_CONSTANT, value=0)


plt.subplot(231), plt.imshow(img, 'gray'), plt.title('ORIGINAL')
plt.subplot(232), plt.imshow(replicate, 'gray'), plt.title('REPLICATE')
plt.subplot(233), plt.imshow(reflect, 'gray'), plt.title('REFLECT')
plt.subplot(234), plt.imshow(reflect101, 'gray'), plt.title('REFLECT_101')
plt.subplot(235), plt.imshow(wrap, 'gray'), plt.title('WRAP')
plt.subplot(236), plt.imshow(constant, 'gray'), plt.title('CONSTANT')

plt.show()
```
