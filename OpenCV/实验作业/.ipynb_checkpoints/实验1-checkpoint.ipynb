```python
import cv2
```

```python
car = cv2.imread('car.png')
car
```

```python
cv2.imshow('car',car)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

```python
carhui = cv2.cvtColor(car,cv2.COLOR_RGB2GRAY) #调用cvtColor函数 转换为灰度图
```

```python
cv2.imshow('car',carhui)
cv2.waitKey(3000) 
cv2.destroyAllWindows()
```

```python
cv2.imwrite('test_image.jpg',carhui) #保存灰度图
```
